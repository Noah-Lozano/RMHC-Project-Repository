```python
import pandas as pd
import matplotlib.pyplot as plt

def filter_2025(df):
    
    # Filter out 2025 data
    df_filtered = df.copy()
    df_filtered["Active Years"] = df_filtered["Active Years"].apply(
        lambda years: [y for y in years if y != 2025]
    )
    return df_filtered

def calculate_churn_rate(df):
    """
    Calculate churn rate for each year.
    Churn rate = (Donors who stopped donating) / (Total donors from previous year)
    """
    churn_data = []
    

    
    # Get all unique years from the dataset
    allYears = []
    for years in df["Active Years"]:
        allYears.extend(years)
    unique = sorted(list(set(allYears)))
    
    for i, year in enumerate(unique):
        if i == 0 :  # First year - no churn rate possible
            active_donors = sum(1 for years in df["Active Years"] if year in years)
            churn_data.append({
                'Year': year,
                'Active_Donors': active_donors,
                'Churn_Rate': 0.0,
                'New_Donors': active_donors,
                'Churned_Donors': 0
            })
        else:
            previous = unique[i-1]
            
            # Count donors active in previous year
            beginning = sum(1 for years in df["Active Years"] if previous in years)
            
            # Count donors active in current year
            end = sum(1 for years in df["Active Years"] if year in years)
            
            
            # Count new donors this year
            new_donors = sum(1 for years in df["Active Years"] 
                           if year in years and previous not in years)
            
            # Calculate churn rate (avoid division by zero)
            churn= ((beginning-end) / beginning) 
            retention = ((end/beginning) * 100)
            
            churn_data.append({
                'Year': year,
                'Active_Donors': end,
                'Churn_Rate': round(churn, 2),
                'New_Donors': new_donors,
                'Retention': retention,
                'Churned_Donors': churn
            })
    
    return churn_data

# Load data
for i in ["inkind", "companies", "people"]:
    df = pd.read_csv(f"donor_summary_{i}.csv")
    print("**************************************************")
    print(i)
    print("**************************************************")
    # Convert Active Years to list of years for each donor

    df["Active Years"] = df["Active Years"].str.split(", ").apply(
    lambda x: [int(y) for y in x if y.isdigit()])
    df=filter_2025(df)
    # Calculate churn rates
    churn_results = calculate_churn_rate(df)

    # Convert to DataFrame for better display
    churn_df = pd.DataFrame(churn_results)
    print("=== CHURN RATE ANALYSIS ===")
    print(churn_df)

    print("=== Percent Increase ===")
    for i in range(len(churn_df)):
        if i == 0:
            print(f"Year: {churn_df['Year'][i]}")
            print(f"Percent Increase: {churn_df['Churn_Rate'][i]}")
        else:
            print(f"From {churn_df['Year'][i-1]} to {churn_df['Year'][i]}")
            print(f"Percent Increase: {((churn_df['Churn_Rate'][i] - churn_df['Churn_Rate'][i-1])/churn_df['Churn_Rate'][i-1] *100) : .2f}%")

    # Create visualizations
    plt.figure(figsize=(15, 10))

    # Plot 1: Churn Rate Over Time
    plt.subplot(2, 3, 1)
    plt.plot(churn_df['Year'], churn_df['Churn_Rate'], marker='o', linewidth=2, markersize=6, color='red')
    plt.title(f'Donor Churn Rate by Year', fontsize=12, fontweight='bold')
    plt.xlabel('Year')
    plt.ylabel('Churn Rate (%)')
    plt.grid(True, alpha=0.3)

    # Plot 2: Active Donors Over Time
    plt.subplot(2, 3, 2)
    plt.plot(churn_df['Year'], churn_df['Active_Donors'], marker='s', linewidth=2, markersize=6, color='green')
    plt.title(f'Active Donors by Year', fontsize=12, fontweight='bold')
    plt.xlabel('Year')
    plt.ylabel('Number of Active Donors')
    plt.grid(True, alpha=0.3)


    # Plot 4: Retention Rate
    plt.subplot(2, 3, 4)
    plt.plot(churn_df['Year'], churn_df['Retention'], marker='^', linewidth=2, markersize=6, color='orange')
    plt.title(f'Donor Retention Rate by Year', fontsize=12, fontweight='bold')
    plt.xlabel('Year')
    plt.ylabel('Retention Rate (%)')
    plt.grid(True, alpha=0.3)

    # Plot 5: Net Growth
    
    #plt.subplot(2, 3, 5)
    #net_growth = churn_df['New_Donors'] - churn_df['Churned_Donors']
    #plt.bar(churn_df['Year'], net_growth, alpha=0.7, color='purple')
    #plt.title(f'Net Donor Growth by Year', fontsize=12, fontweight='bold')
    #plt.xlabel('Year')
    #plt.ylabel('Net Growth (New - Churned)')
    #plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)


    plt.tight_layout()
    plt.show()

    # Print summary statistics
    print("\n=== CHURN RATE SUMMARY ===")
    print(f"Average Churn Rate: {churn_df['Churn_Rate'].mean():.2f}%")
    print(f"Average Retention Rate: {churn_df['Retention'].mean():.2f}%")
    print(f"Year with Highest Churn: {churn_df.loc[churn_df['Churn_Rate'].idxmax(), 'Year']} ({churn_df['Churn_Rate'].max():.2f}%)")
    print(f"Year with Lowest Churn: {churn_df.loc[churn_df['Churn_Rate'].idxmin(), 'Year']} ({churn_df['Churn_Rate'].min():.2f}%)")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def count_donations_by_year(df, source_name):
    print(f"\n{'='*50}")
    print(f"ANALYZING: {source_name}")
    print(f"{'='*50}")
    
    # Convert Entry Date to datetime
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df['Year'] = df['Date'].dt.year
    
    # Filter out 2025 data
    df = df[df['Year'] != 2025]
    
    # Count donations by year
    donations_by_year = df['Year'].value_counts().sort_index()
    
    print(f"📊 Total donations: {len(df)}")
    print(f"📅 Years with data: {sorted(df['Year'].unique())}")
    print(f"👥 Unique donors: {df['Entity #'].nunique()}")
    
    print(f"\n📈 Donations by Year:")
    for year, count in donations_by_year.items():
        print(f"   {year}: {count} donations")
    
    return donations_by_year


def analyze_donation_counts():
    # Load the data
    print("Loading data files...")
    inkind = pd.read_csv("RMHC Inkind.csv")
    companies = pd.read_csv("RMHC Companies.csv")
    households = pd.read_csv("RMHC Households.csv")
    
    # Process each dataset
    inkind_counts = count_donations_by_year(inkind, "Inkind Donations")
    companies_counts = count_donations_by_year(companies, "Company Donations")
    households_counts = count_donations_by_year(households, "Household Donations")
    
    # Create combined dataframe for analysis
    all_years = sorted(set(list(inkind_counts.index) + list(companies_counts.index) + list(households_counts.index)))
    
    combined_data = pd.DataFrame({
        'Year': all_years,
        'Inkind': [inkind_counts.get(year, 0) for year in all_years],
        'Companies': [companies_counts.get(year, 0) for year in all_years],
        'Households': [households_counts.get(year, 0) for year in all_years]
    })
    
    # Calculate totals
    combined_data['Total'] = combined_data['Inkind'] + combined_data['Companies'] + combined_data['Households']
    
    print(f"\n{'='*60}")
    print("COMBINED DONATION COUNTS BY YEAR")
    print(f"{'='*60}")
    print(combined_data.to_string(index=False))  
    return combined_data

def create_visualizations(donation_data):
    """
    Create visualizations for donation counts
    """
    plt.figure(figsize=(15, 10))
    
    # Plot 1: Donations by year for each source
    plt.subplot(2, 2, 1)
    plt.plot(donation_data['Year'], donation_data['Inkind'], marker='o', label='Inkind', linewidth=2, markersize=6, color= 'blue')
    plt.plot(donation_data['Year'], donation_data['Companies'], marker='s', label='Companies', linewidth=2, markersize=6, color= 'orange')
    plt.plot(donation_data['Year'], donation_data['Households'], marker='^', label='Households', linewidth=2, markersize=6, color= 'green')
    plt.title('Donations by Year - Each Source', fontsize=12, fontweight='bold')
    plt.xlabel('Year')
    plt.ylabel('Number of Donations')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Plot 2: Total donations by year
    plt.subplot(2, 2, 2)
    plt.bar(donation_data['Year'], donation_data['Total'], alpha=0.7, color='green')
    plt.title('Total Donations by Year', fontsize=12, fontweight='bold')
    plt.xlabel('Year')
    plt.ylabel('Total Number of Donations')
    plt.grid(True, alpha=0.3)
    
    # Plot 3: Stacked bar chart
    plt.subplot(2, 2, 3)
    plt.bar(donation_data['Year'], donation_data['Inkind'], label='Inkind', alpha=0.7, color= 'blue')
    plt.bar(donation_data['Year'], donation_data['Companies'], bottom=donation_data['Inkind'], label='Companies', alpha=0.7, color= 'orange')
    plt.bar(donation_data['Year'], donation_data['Households'], bottom=donation_data['Inkind']+donation_data['Companies'], label='Households', alpha=0.7, color= 'green')
    plt.title('Donations by Year Stacked Bar Chart', fontsize=12, fontweight='bold')
    plt.xlabel('Year')
    plt.ylabel('Number of Donations')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Plot 4: Donation trends
    plt.subplot(2, 2, 4)
    plt.plot(donation_data['Year'], donation_data['Total'], marker='o', linewidth=3, color='red', markersize=8)
    plt.title('Overall Donations', fontsize=12, fontweight='bold')
    plt.xlabel('Year')
    plt.ylabel('Total Donations')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def print_summary_statistics(donation_data):
    """
    Print summary statistics
    """
    print("--------------------------------")
    print("SUMMARY OF DONATIONS")
    print("--------------------------------")
    print(f"Total donations across all types: {donation_data['Total'].sum():,}")
    print(f"Highest year: {donation_data.loc[donation_data['Total'].idxmax(), 'Year']} ({donation_data['Total'].max():,} donations)")
    print(f"Lowest year: {donation_data.loc[donation_data['Total'].idxmin(), 'Year']} ({donation_data['Total'].min():,} donations)")
    #print(f"Average donations per year: {donation_data['Total'].mean():.0f}")
    
    # Show which source contributes most
    total_by_source = {
        'Inkind': donation_data['Inkind'].sum(),
        'Companies': donation_data['Companies'].sum(),
        'Households': donation_data['Households'].sum()
    }
    
    print(f"\nDonor Types:")
    for source, count in sorted(total_by_source.items(), key=lambda x: x[1], reverse=True):
        percentage = (count / sum(total_by_source.values())) * 100
        print(f"{source}: {count:,} donations ({percentage:.1f}%)")

# Main execution
if __name__ == "__main__":
    # Analyze donation counts
    donation_data = analyze_donation_counts()
    
    # Create visualizations
    create_visualizations(donation_data)
    
    # Print summary
    print_summary_statistics(donation_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def analyze_2022_monthly():
    """
    Simple 2022 monthly analysis for households and companies
    """
    print("Loading 2022 data...")
    
    # Load data
    companies = pd.read_csv("RMHC Companies.csv")
    households = pd.read_csv("RMHC Households.csv")
    
    # Convert dates and filter for 2022
    companies['Date'] = pd.to_datetime(companies['Date'], errors='coerce')
    households['Date'] = pd.to_datetime(households['Date'], errors='coerce')
    
    companies_2022 = companies[companies['Date'].dt.year == 2021].copy()
    households_2022 = households[households['Date'].dt.year == 2021].copy()
    
    # Add month info
    companies_2022['Month'] = companies_2022['Date'].dt.month
    households_2022['Month'] = households_2022['Date'].dt.month
    
    # Clean monetary data
    companies_2022['Gift Amount'] = companies_2022['Gift Amount'].str.replace('$', '').str.replace(',', '').str.replace('(', '').str.replace(')', '').astype(float)
    households_2022['Gift Amount'] = households_2022['Gift Amount'].str.replace('$', '').str.replace(',', '').str.replace('(', '').str.replace(')', '').astype(float)
    
    # Monthly analysis
    companies_monthly = companies_2022.groupby('Month').agg({
        'Gift Amount': ['sum', 'count']
    }).reset_index()
    companies_monthly.columns = ['Month', 'Total_Amount', 'Count']
    
    households_monthly = households_2022.groupby('Month').agg({
        'Gift Amount': ['sum', 'count']
    }).reset_index()
    households_monthly.columns = ['Month', 'Total_Amount', 'Count']
    
    # Create visualizations
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
    
    # 1. Donation Counts - Bar Graph
    x = range(1, 13)
    width = 0.35
    
    ax1.bar([i - width/2 for i in x], companies_monthly['Count'], width, label='Companies', alpha=0.8)
    ax1.bar([i + width/2 for i in x], households_monthly['Count'], width, label='Households', alpha=0.8)
    ax1.set_title('2021 Monthly Donation Counts', fontweight='bold')
    ax1.set_xlabel('Month')
    ax1.set_ylabel('Number of Donations')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # 2. Donation Counts - Line Graph
    ax2.plot(companies_monthly['Month'], companies_monthly['Count'], marker='o', label='Companies', linewidth=2)
    ax2.plot(households_monthly['Month'], households_monthly['Count'], marker='s', label='Households', linewidth=2)
    ax2.set_title('2021 Monthly Donation Counts', fontweight='bold')
    ax2.set_xlabel('Month')
    ax2.set_ylabel('Number of Donations')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    # 3. Monetary Amounts - Bar Graph
    ax3.bar([i - width/2 for i in x], companies_monthly['Total_Amount'], width, label='Companies', alpha=0.8)
    ax3.bar([i + width/2 for i in x], households_monthly['Total_Amount'], width, label='Households', alpha=0.8)
    ax3.set_title('2021 Monthly Donation Amounts', fontweight='bold')
    ax3.set_xlabel('Month')
    ax3.set_ylabel('Total Amount ($)')
    ax3.legend()
    ax3.grid(True, alpha=0.3)
    
    # 4. Monetary Amounts - Line Graph
    ax4.plot(companies_monthly['Month'], companies_monthly['Total_Amount'], marker='o', label='Companies', linewidth=2)
    ax4.plot(households_monthly['Month'], households_monthly['Total_Amount'], marker='s', label='Households', linewidth=2)
    ax4.set_title('2021 Monthly Donation Amounts', fontweight='bold')
    ax4.set_xlabel('Month')
    ax4.set_ylabel('Total Amount ($)')
    ax4.legend()
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Print summary
    print(f"\n2021 Summary:")
    print(f"Companies: {companies_monthly['Count'].sum():,} donations, ${companies_monthly['Total_Amount'].sum():,.2f}")
    print(f"Households: {households_monthly['Count'].sum():,} donations, ${households_monthly['Total_Amount'].sum():,.2f}")

if __name__ == "__main__":
    analyze_2022_monthly() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def calculate_churn_retention(df, source_name):
    """
    Calculate churn and retention rates for a dataset
    """
    print(f"\n{'='*60}")
    print(f"CHURN & RETENTION ANALYSIS: {source_name}")
    print(f"{'='*60}")
    
    # Convert Entry Date to datetime
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df['Year'] = df['Date'].dt.year
    
    # Filter out 2025 data
    df = df[df['Year'] != 2025]
    
    # Get active years for each donor
    active_years = df.groupby('Donor Name')['Year'].unique().reset_index()
    active_years['Active Years'] = active_years['Year'].apply(lambda x: sorted(x))
    
    # Calculate churn data
    churn_data = []
    all_years = sorted(df['Year'].unique())
    
    for i, year in enumerate(all_years):
        if i == 0:  # First year - no churn rate possible
            active_donors = sum(1 for years in active_years['Active Years'] if year in years)
            churn_data.append({
                'Year': year,
                'Active_Donors': active_donors,
                'Churn_Rate': 0.0,
                'Retention_Rate': 100.0,
                'New_Donors': active_donors,
                'Churned_Donors': 0
            })
        else:
            previous_year = all_years[i-1]
            
            # Count donors active in previous year
            beginning_donors = sum(1 for years in active_years['Active Years'] if previous_year in years)
            
            # Count donors active in current year
            end_donors = sum(1 for years in active_years['Active Years'] if year in years)
            
            # Count new donors this year
            new_donors = sum(1 for years in active_years['Active Years'] 
                           if year in years and previous_year not in years)
            
            # Calculate churn and retention rates
            if beginning_donors > 0:
                churn_rate = ((beginning_donors - end_donors) / beginning_donors) * 100
                retention_rate = (end_donors / beginning_donors) * 100
            else:
                churn_rate = 0
                retention_rate = 100
            
            churn_data.append({
                'Year': year,
                'Active_Donors': end_donors,
                'Churn_Rate': round(churn_rate, 2),
                'Retention_Rate': round(retention_rate, 2),
                'New_Donors': new_donors,
                'Churned_Donors': beginning_donors - end_donors
            })
    
    return pd.DataFrame(churn_data)

def create_simple_visualizations(churn_df, source_name):
    """
    Create simple visualizations with numbers on points
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # 1. Churn Rate Over Time
    ax1.plot(churn_df['Year'], churn_df['Churn_Rate'], marker='o', linewidth=3, markersize=8, color='red')
    ax1.set_title(f'Churn Rate - {source_name}', fontsize=14, fontweight='bold')
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Churn Rate (%)')
    ax1.grid(True, alpha=0.3)
    
    # Add numbers on points
    for x, y in zip(churn_df['Year'], churn_df['Churn_Rate']):
        ax1.annotate(f'{y}%', (x, y), textcoords="offset points", xytext=(0,10), ha='center', fontweight='bold')
    
    # 2. Retention Rate Over Time
    ax2.plot(churn_df['Year'], churn_df['Retention_Rate'], marker='s', linewidth=3, markersize=8, color='green')
    ax2.set_title(f'Retention Rate - {source_name}', fontsize=14, fontweight='bold')
    ax2.set_xlabel('Year')
    ax2.set_ylabel('Retention Rate (%)')
    ax2.grid(True, alpha=0.3)
    
    # Add numbers on points
    for x, y in zip(churn_df['Year'], churn_df['Retention_Rate']):
        ax2.annotate(f'{y}%', (x, y), textcoords="offset points", xytext=(0,10), ha='center', fontweight='bold')
    
    plt.tight_layout()
    plt.show()

def print_detailed_data(churn_df, source_name):
    """
    Print all data in a clear format
    """
    print(f"\nDETAILED DATA TABLE - {source_name}")
    print(f"{'='*60}")
    print(churn_df.to_string(index=False))
    
    print(f"\nSUMMARY STATISTICS - {source_name}")
    print(f"{'='*50}")
    print(f"Average Churn Rate: {churn_df['Churn_Rate'].mean():.2f}%")
    print(f"Average Retention Rate: {churn_df['Retention_Rate'].mean():.2f}%")
    print(f"Highest Churn: {churn_df.loc[churn_df['Churn_Rate'].idxmax(), 'Year']} ({churn_df['Churn_Rate'].max():.2f}%)")
    print(f"Lowest Churn: {churn_df.loc[churn_df['Churn_Rate'].idxmin(), 'Year']} ({churn_df['Churn_Rate'].min():.2f}%)")
    
    print(f"\nEAR-BY-YEAR BREAKDOWN - {source_name}")
    print(f"{'='*50}")
    for _, row in churn_df.iterrows():
        print(f"Year {row['Year']}:")
        print(f"  Active Donors: {row['Active_Donors']:,}")
        print(f"  Churn Rate: {row['Churn_Rate']:.2f}%")
        print(f"  Retention Rate: {row['Retention_Rate']:.2f}%")
        print(f"  New Donors: {row['New_Donors']:,}")
        print(f"  Churned Donors: {row['Churned_Donors']:,}")
        print()

def analyze_all_sources():
    """
    Analyze churn and retention for all three data sources
    """
    print("Loading and analyzing all data sources...")
    
    # Load data
    inkind = pd.read_csv("RMHC Inkind.csv")
    companies = pd.read_csv("RMHC Companies.csv")
    households = pd.read_csv("RMHC Households.csv")
    
    # Analyze each source
    sources = [
        (inkind, "Inkind Donations"),
        (companies, "Company Donations"),
        (households, "Household Donations")
    ]
    
    all_results = {}
    
    for df, source_name in sources:
        print(f"\n{'='*80}")
        print(f"ANALYZING: {source_name}")
        print(f"{'='*80}")
        
        # Calculate churn and retention
        churn_df = calculate_churn_retention(df, source_name)
        all_results[source_name] = churn_df
        
        # Print detailed data
        print_detailed_data(churn_df, source_name)
        
        # Create simple visualizations
        create_simple_visualizations(churn_df, source_name)
    
    # Print comparison summary
    print_comparison_summary(all_results)
    
    return all_results

def print_comparison_summary(all_results):
    """
    Print comparison summary across all sources
    """
    print(f"\n{'='*80}")
    print("CROSS-SOURCE COMPARISON SUMMARY")
    print(f"{'='*80}")
    
    print(f"{'Source':<20} {'Avg Churn':<12} {'Avg Retention':<15} {'Peak Year':<12} {'Peak Churn':<12}")
    print("-" * 80)
    
    for source_name, churn_df in all_results.items():
        avg_churn = churn_df['Churn_Rate'].mean()
        avg_retention = churn_df['Retention_Rate'].mean()
        peak_year = churn_df.loc[churn_df['Churn_Rate'].idxmax(), 'Year']
        peak_churn = churn_df['Churn_Rate'].max()
        
        print(f"{source_name.split()[0]:<20} {avg_churn:<12.2f} {avg_retention:<15.2f} {peak_year:<12} {peak_churn:<12.2f}")

# Main execution
if __name__ == "__main__":
    # Run analysis for all sources
    results = analyze_all_sources()
```
