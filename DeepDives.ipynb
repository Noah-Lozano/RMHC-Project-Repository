```python
import pandas as pd
import matplotlib.pyplot as plt

def analyze_year_monthly():
    for i in 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025:
        print(str(i)+ "Data")
        companies = pd.read_csv("RMHC Companies.csv")
        households = pd.read_csv("RMHC Households.csv")
        companies = companies[companies['Donor Name'] != 'AbbVie, Inc.']
        
        # Convert dates and filter for 2022
        companies['Date'] = pd.to_datetime(companies['Date'], errors='coerce')
        households['Date'] = pd.to_datetime(households['Date'], errors='coerce')
        companies_i = companies[companies['Date'].dt.year == i].copy()
        households_i = households[households['Date'].dt.year == i].copy()
        
        # Add month info
        companies_i['Month'] = companies_i['Date'].dt.month
        households_i['Month'] = households_i['Date'].dt.month

        companies_monthly = companies_i.groupby('Month').agg({
            'Gift Value': ['sum', 'count']
        }).reset_index()
        companies_monthly.columns = ['Month', 'Total_Amount', 'Count']
        
        households_monthly = households_i.groupby('Month').agg({
            'Gift Value': ['sum', 'count']
        }).reset_index()
        households_monthly.columns = ['Month', 'Total_Amount', 'Count']

        all_months = pd.DataFrame({'Month': range(1, 13)})
        
        companies_monthly = all_months.merge(companies_monthly, on='Month', how='left').fillna(0)
        households_monthly = all_months.merge(households_monthly, on='Month', how='left').fillna(0)
        
        # Create visualizations
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
        
        # 1. Donation Counts - Bar Graph
        x = range(1, 13)
        width = 0.35
        
        ax1.bar([i - width/2 for i in x], companies_monthly['Count'], width, label='Companies', alpha=0.8)
        ax1.bar([i + width/2 for i in x], households_monthly['Count'], width, label='Households', alpha=0.8)
        ax1.set_title(str(i)+' Monthly Donation Counts', fontweight='bold')
        ax1.set_xlabel('Month')
        ax1.set_ylabel('Number of Donations')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # 2. Donation Counts - Line Graph
        ax2.plot(companies_monthly['Month'], companies_monthly['Count'], marker='o', label='Companies', linewidth=2)
        ax2.plot(households_monthly['Month'], households_monthly['Count'], marker='s', label='Households', linewidth=2)
        ax2.set_title(str(i)+' Monthly Donation Counts', fontweight='bold')
        ax2.set_xlabel('Month')
        ax2.set_ylabel('Number of Donations')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        
        # 3. Monetary Amounts - Bar Graph
        ax3.bar([i - width/2 for i in x], companies_monthly['Total_Amount'], width, label='Companies', alpha=0.8)
        ax3.bar([i + width/2 for i in x], households_monthly['Total_Amount'], width, label='Households', alpha=0.8)
        ax3.set_title(str(i)+' Monthly Donation Amounts', fontweight='bold')
        ax3.set_xlabel('Month')
        ax3.set_ylabel('Total Amount ($)')
        ax3.legend()
        ax3.grid(True, alpha=0.3)
        
        # 4. Monetary Amounts - Line Graph
        ax4.plot(companies_monthly['Month'], companies_monthly['Total_Amount'], marker='o', label='Companies', linewidth=2)
        ax4.plot(households_monthly['Month'], households_monthly['Total_Amount'], marker='s', label='Households', linewidth=2)
        ax4.set_title(str(i)+' Monthly Donation Amounts', fontweight='bold')
        ax4.set_xlabel('Month')
        ax4.set_ylabel('Total Amount ($)')
        ax4.legend()
        ax4.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
        # Print monthly breakdown
        print(f"\nMonthly Breakdown for {i}:")
        print(f"{'Month':<8} {'Companies':<12} {'Households':<12} {'Total':<12}")
        print("-" * 50)
        for month in range(1, 13):
            comp_count = companies_monthly[companies_monthly['Month'] == month]['Count'].iloc[0]
            house_count = households_monthly[households_monthly['Month'] == month]['Count'].iloc[0]
            total = comp_count + house_count
            comp_money_count= companies_monthly[companies_monthly['Month'] == month]['Total_Amount'].iloc[0]
            house_money_count= households_monthly[households_monthly['Month'] == month]['Total_Amount'].iloc[0]
            money_total =comp_money_count + house_money_count
            print(f"{month:<8} {comp_count:<12} {house_count:<12} {total:<12}  {comp_money_count:<12} {house_money_count:<12} {money_total:<12}")
        
        # Print summary
        print(f"\n" + str(i) + " Summary:")
        print(f"Companies: {companies_monthly['Count'].sum():,} donations, ${companies_monthly['Total_Amount'].sum():,.2f}")
        print(f"Households: {households_monthly['Count'].sum():,} donations, ${households_monthly['Total_Amount'].sum():,.2f}")

if __name__ == "__main__":
    analyze_year_monthly() 
```
